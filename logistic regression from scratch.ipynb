```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=7)
```

```python
model = LogisticRegression()
model.fit(x_test,y_test)
pred=model.predict(x_test)
```

```python
def sigmoid(x):
    return 1/(1+np.exp(x))
def output(x):
    return np.where(x<0.5,0,1)
```

```python
class logistic:
    def __init__(self,lr=0.1,epochs=100):
        self.coef=None
        self.intercept=None
        self.lr=lr
        self.epoch=epochs
    def fit(self,x_train,y_train):
        self.intercept=0
        x_train=np.insert(x_train,0,1,axis=1)
        self.coef=np.ones(x_train.shape[1])
        for i in range(self.epoch):
            y_hat=sigmoid(np.dot(x_train,self.coef))
            slope = np.dot((y_train-y_hat),x_train)/x_train.shape[0]
            self.coef = self.coef - (self.lr*slope)
        self.intercept=self.coef[0]
        self.coef = self.coef[1:]

    def predict(self,x_test):
        return output(sigmoid(np.dot(x_test,self.coef)))
```

```python
mera=logistic()
mera.fit(x_train,y_train)
mera_pred=mera.predict(x_test)
```
